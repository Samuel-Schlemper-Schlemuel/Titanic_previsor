# Cleaning the DataBase

One or other thing I got of the [MUHAMMAD IBRAHIM QASMI](https://www.kaggle.com/code/muhammadibrahimqasmi/titanic-survival-percentage-beginner-friendly)

```python
raw_train = pd.read_csv('train.csv')
raw_test = pd.read_csv('test.csv')
```

```python
raw_train.head()
```

```python
raw_train.info()
```

```python
raw_train.describe()
```

```python
raw_train = raw_train.drop(['Ticket', 'Name', 'PassengerId'], axis=1)
raw_test = raw_test.drop(['Ticket', 'Name'], axis=1)
```

```python
print(raw_train.isnull().sum())
print('---------------------')
print(raw_test.isnull().sum())
```

```python
raw_train['Age Input'] = raw_train['Age'].apply(lambda x: 1 if pd.isna(x) else 0)
raw_train['Age'] = raw_train['Age'].fillna(raw_train['Age'].mean())
raw_train['Cabin'] = raw_train['Cabin'].fillna(0)

raw_test['Age Input'] = raw_test['Age'].apply(lambda x: 1 if pd.isna(x) else 0)
raw_test['Age'] = raw_test['Age'].fillna(raw_test['Age'].mean())
raw_test['Fare'] = raw_test['Fare'].fillna(raw_test['Fare'].mean())
raw_test['Cabin'] = raw_test['Cabin'].fillna(0)
```

```python
print(raw_train.isnull().sum())
print('---------------------')
print(raw_test.isnull().sum())
```

```python
raw_train = pd.get_dummies(raw_train, columns=["Sex", "Embarked", "Pclass"])
raw_test = pd.get_dummies(raw_test, columns=["Sex", "Embarked", "Pclass"])
```

```python
raw_train['SibSp'] = raw_train['SibSp'].apply(lambda x: 0 if x == 0 else 1)
raw_train['Parch'] = raw_train['Parch'].apply(lambda x: 0 if x == 0 else 1)
raw_train['Cabin'] = raw_train['Cabin'].apply(lambda x: 0 if x == 0 else 1)

raw_test['SibSp'] = raw_test['SibSp'].apply(lambda x: 0 if x == 0 else 1)
raw_test['Parch'] = raw_test['Parch'].apply(lambda x: 0 if x == 0 else 1)
raw_test['Cabin'] = raw_test['Cabin'].apply(lambda x: 0 if x == 0 else 1)
```

```python
x_train = raw_train.drop("Survived", axis=1)
y_train = raw_train["Survived"]
```

```python
x_train.head()
```

# Making the AI

#### With validation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def testing():
  all = []

  options = []

  for n in range(90, 321, 10):
    for depth in range(5, 19):
      for split in range(2, 5):
        options.append([n, depth, split])

  for op in options:
    sum = 0

    for num in range(0, 5):
      train_x, val_x, train_y, val_y = train_test_split(x_train, y_train, random_state = num, test_size = 0.2)
      model = RandomForestClassifier(random_state = 0, n_estimators = op[0], criterion = 'entropy', max_depth = op[1], min_samples_split = op[2])
      model.fit(train_x, train_y)
      preds = model.predict(val_x).round()
      sum += round(100 * accuracy_score(val_y, preds), 2)

    all.append({f'{op}': round(sum/5, 2)})

  return all
```

```python
ar = testing()
print(ar)
```

```python
max = {'irrelevant': 0}

for obj in ar:
  if list(obj.values())[0] > list(max.values())[0]:
    max = obj

print(max)
```
